### Analysis

- Ten different Drugs were part of this study, each with a minimum of 125 treatments to mice. 
- of the Drugs Capomulin, Ramicane, Infubinol and Ceftamin, Capomulin and Ramicane resulted in samller End Tumor Volume ranges (20 to high 40s mm3 v. Infubinol and Ceftamin's 40s to 70+ mm3). 
- For the Capomulin drug, the Average Tumor volume increased with the weight of the test mouse, indicating weight influences the outcome of treatment. 
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import pearsonr, linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merge_study_data=pd.merge(mouse_metadata, study_results, on="Mouse ID")
merge_study_data

# Display the data table for preview
merge_study_data.head()
```

```python
# Checking the number of mice.
mice_count=merge_study_data["Mouse ID"].value_counts()
mice_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = merge_study_data[merge_study_data.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)]
duplicate_ids = duplicate_mice['Mouse ID'].unique()
print("Duplicate Mouse IDs")
print(duplicate_ids)
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_data = merge_study_data.loc[merge_study_data['Mouse ID'].isin(duplicate_ids)]
duplicate_data.head()
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
merge_study_data_clean = merge_study_data[~merge_study_data["Mouse ID"].isin(duplicate_ids)]
merge_study_data_clean.head()
```

```python
# Checking the number of mice in the clean DataFrame.
mice_count=merge_study_data_clean["Mouse ID"].value_counts()
mice_count
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
Tumor_Volume=merge_study_data[["Drug Regimen", "Mouse ID", "Timepoint", "Tumor Volume (mm3)"]]
Tumor_Volume

drug_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Filter data for the specified treatment regimens
filtered_tumor_volume = merge_study_data[merge_study_data["Drug Regimen"].isin(drug_list)]

# Get the last (greatest) timepoint for each mouse across specified treatment regimens
last_timepoints = filtered_tumor_volume.groupby("Mouse ID")["Timepoint"].max().reset_index()

# Merge to get the final tumor volume at the last timepoint for each mouse
final_tumor_volume = pd.merge(last_timepoints, merge_study_data, on=["Mouse ID", "Timepoint"], how="left")

# Display the final tumor volume for each mouse across the specified treatment regimens
final_tumor_volume = final_tumor_volume[["Mouse ID", "Drug Regimen", "Timepoint", "Tumor Volume (mm3)"]]

final_tumor_volume = final_tumor_volume.rename(columns={"Tumor Volume (mm3)": "End Tumor Volume"})
final_tumor_volume

#Merge this group df with the orginal DataFrame to get the tumor volume at the last timepoint
end_merge_study_data=pd.merge(final_tumor_volume, merge_study_data, on=["Mouse ID", "Timepoint", "Drug Regimen"],how="left")
end_merge_study_data
end_merge_study_data.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drug_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_volume=[]

for drug in drug_list:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_vol_data = end_merge_study_data.loc[end_merge_study_data['Drug Regimen'] == drug, 'End Tumor Volume']
    
    # Append tumor volume data to the list
    tumor_volume.append(tumor_vol_data)
    
    # Calculate quartiles and IQR
    quartiles = tumor_vol_data.quantile([0.25, 0.75])
    lower_quartile = quartiles[0.25]
    upper_quartile = quartiles[0.75]
    iqr = upper_quartile - lower_quartile
    
    # Determine outliers using upper and lower bounds
    lower_bound = lower_quartile - (1.5 * iqr)
    upper_bound = upper_quartile + (1.5 * iqr)
    
    # Display potential outliers
    outliers = tumor_vol_data.loc[(tumor_vol_data < lower_bound) | (tumor_vol_data > upper_bound)]
    print(f"Potential outliers for {drug}: {outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
flierprops=dict(marker="o", markerfacecolor="red", markersize=12, markeredgecolor="none")

plt.figure(figsize=(8, 6))
boxplot = plt.boxplot(tumor_volume, labels=drug_list, flierprops=flierprops)

plt.title('Distribution of End Tumor Volume for Each Treatment Group')
plt.xlabel('Drug Treatment')
plt.ylabel('End Tumor Volume')
plt.grid(False)

plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin

capomulin_data = merge_study_data[(merge_study_data['Drug Regimen'] == 'Capomulin') & (merge_study_data['Mouse ID'] == 'l509')]

# Create a line plot for tumor volume vs. time point
plt.figure(figsize=(8, 6))
plt.plot(capomulin_data['Timepoint'], capomulin_data['Tumor Volume (mm3)'], color='blue')

plt.title('Capomulin treatment for mouse l509')
plt.xlabel('Time Point (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.grid(False)

plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

# Filter data for the Capomulin regimen
capomulin_data = merge_study_data[merge_study_data['Drug Regimen'] == 'Capomulin']

# Calculate average tumor volume for each mouse
average_tumor_volume = capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()

# Obtain corresponding weights for the mice in the Capomulin regimen
mouse_weights = capomulin_data.groupby('Mouse ID')['Weight (g)'].mean()

# Create a scatter plot of mouse weight vs. average tumor volume
plt.figure(figsize=(8, 6))
plt.scatter(mouse_weights, average_tumor_volume, marker='o', color='blue')

plt.title('Mouse Weight vs. Average Tumor Volume (Capomulin Regimen)')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.grid(False)

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

# Calculate correlation coefficient using pearsonr
correlation=st.pearsonr(mouse_weights, average_tumor_volume)

print(f'The correlation between Mouse Weight and Average Tumor Volume is {round(correlation[0],2)}')

# Performing linear regression
slope, intercept, r_value, p_value, std_err = linregress(mouse_weights, average_tumor_volume)

# Plotting the data and regression line
plt.figure(figsize=(8, 6))
plt.scatter(mouse_weights, average_tumor_volume, color='blue', label='Data points')
plt.plot(mouse_weights, slope * mouse_weights + intercept, color='red', label='Regression Line')

plt.title('Mouse Weight vs. Average Tumor Volume (Capomulin Regimen)')
plt.xlabel('Mouse Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.grid(False)
plt.show()
```
